```python
import numpy as np
import matplotlib.pyplot as plt

def wave_function(x, t, a, n):
    hbar = 1
    m = 1
    psi = (8 * np.sqrt(30) / (np.pi ** 3 * np.sqrt(a))) * np.sum(
        (1 / n**3) * np.sin((n * np.pi /a)* x) * np.exp(-1j * n**2 * np.pi**2 * hbar * t / (2 * m * a**2))
        for n in range(1, n + 1, 2))
    return psi

a = 1
t = 1
x = np.linspace(-a, a, 1000)

plt.figure(figsize=(10, 6))
for n in range(1, 10, 2):
    psi = wave_function(x, t, a, n)
    plt.plot(x, psi.real, label=f"n = {n}")

plt.xlabel("x")
plt.ylabel("(x, t)")
plt.title("Wave Functions")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
for n in range(1, 10, 2):
    psi = wave_function(x, t, a, n)
    probability = np.abs(psi) ** 2
    plt.plot(x, probability, label=f"n = {n}")

plt.xlabel("x")
plt.ylabel("|(x, t)|^2")
plt.title("Probability Distributions")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del problema
a = 1.0
t = 1.0
m = 1.0

# Funcin para calcular los coeficientes c_n
def cn(n, a):
    if n % 2 == 0:
        return 0
    else:
        return (4 * np.sqrt(30) / ((n * np.pi) ** 3)) * np.sin((n * np.pi * a) / (2 * a))

# Funcin para calcular la funcin de onda en un tiempo dado
def psi(x, t, a, n):
    if n % 2 == 0:
        return np.zeros_like(x)
    else:
        psi_n = np.sqrt(2 / a) * np.sin((n * np.pi * x) / a) * np.exp(
            -1j * ((n ** 2 * np.pi ** 2 * t) / (2 * m * a ** 2))
        )
        return psi_n

# Valores de n para los cuales calcular la funcin de onda
n_values = [1, 2, 3, 4, 5]

# Valores de x para el grfico
x = np.linspace(0, a, 500)

for n in n_values:
    # Calcula la funcin de onda para el valor de n dado
    psi_values = psi(x, t, a, n)
    
    # Calcula la distribucin de probabilidad
    psi_prob = np.abs(psi_values) ** 2

    # Grfico de la parte real e imaginaria de la funcin de onda
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(x, np.real(psi_values), label='Parte Real')
    plt.plot(x, np.imag(psi_values), label='Parte Imaginaria')
    plt.title(fr'Re($\psi(x,t)$) e Im($\psi(x,t)$) para $n={n}$')
    plt.xlabel('x')
    plt.ylabel('Re($\psi$), Im($\psi$)')
    plt.legend()

    # Grfico de la distribucin de probabilidad
    plt.subplot(1, 2, 2)
    plt.plot(x, psi_prob, label='Probabilidad')
    plt.title(fr'Distribucin de Probabilidad para $n={n}$')
    plt.xlabel('x')
    plt.ylabel('Probabilidad')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
